## Lemonae Sale Prediction

```python
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```python
# 데이터 준비
data = pd.read_csv('./csv/lemonade.csv')
data.head()
```

```python
# 종속변수, 독립변수
independent = data[['온도']]
subordination = data[['판매량']]
print(independent.shape, subordination.shape)
```

```python
# 모델을 만듭니다.
x = tf.keras.layers.Input(shape=[1])
y = tf.keras.layers.Dense(1)(x)
model = tf.keras.models.Model(x, y)
model.compile(loss='mse')
```

```python
# 모델을 학습합니다.
model.fit(independent, subordination, epochs=10000)
```

```python
# 모델을 이용합니다
model.predict(independent)
```
