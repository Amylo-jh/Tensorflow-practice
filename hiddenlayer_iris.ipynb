```python
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```python
# 1. 과거의 데이터를 준비한다.
iris = pd.read_csv('./csv/iris.csv')
iris.head()
```

```python
# 원핫인코딩
encoding = pd.get_dummies(iris)
encoding.head()
```

```python
encoding.columns
```

```python
# 독립변수, 종속변수 분리
independent = encoding[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
subordination = encoding[['품종_setosa', '품종_versicolor','품종_virginica']]
print(independent.shape, subordination.shape)
```

```python
# 2. 모델의 구조를 만든다.
x = tf.keras.layers.Input(shape=[4])
h = tf.keras.layers.Dense(8, activation='swish')(x)
h = tf.keras.layers.Dense(8, activation='swish')(h)
h = tf.keras.layers.Dense(8, activation='swish')(h)
y = tf.keras.layers.Dense(3, activation='softmax')(h)
model = tf.keras.models.Model(x, y)
model.compile(loss='categorical_crossentropy', metrics='accuracy')
```

```python
model.fit(independent, subordination, epochs=100)
```

```python
print(model.predict(independent[-5:]).round(3))
print(subordination[-5:])
```
