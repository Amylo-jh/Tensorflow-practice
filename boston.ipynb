## 보스턴 집값 예측

```python
# 라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```python
# 1. 과거의 데이터를 준비한다.
boston = pd.read_csv('./csv/boston.csv')
print(boston.head())
print(boston.columns)
```

```python
# 독립변수, 종속변수 분리
independent = boston[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat']]
subordination = boston[['medv']]
print(independent.shape, subordination.shape)
```

```python
# 모델의 구조를 만든다.
x = tf.keras.layers.Input(shape=[13])
y = tf.keras.layers.Dense(1)(x)
model = tf.keras.models.Model(x, y)
model.compile(loss='mse')
```

```python
# 데이터로 모델을 학습(fit)한다.
model.fit(independent, subordination, epochs=1000)
```

```python
# 4. 모델을 이용합니다.
model.predict(independent[0:5])
```
